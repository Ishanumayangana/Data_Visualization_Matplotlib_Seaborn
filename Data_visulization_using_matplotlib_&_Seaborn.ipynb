creating line chart


```python
import matplotlib.pyplot as plt

x=[1,2,3,4,5]
y=[85,89,79,87,81]

plt.plot(x,y)

plt.title("Performance Report")

plt.xlabel("Semester Number")
plt.ylabel("Percentage of marks")
plt.show()
```

creating histogran

```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5,6,7,4]
plt.hist(x,bins=[1,2,3,4,5,6,7])
plt.title("Histogram")
plt.legend(["bar"])
plt.show
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4]

e = (0.1,0,0,0)
plt.pie(x,explode = e)

plt.title("pie chart")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')

sns.violinplot(x='day', y='total_bill', hue='sex', data=data, split=True, palette='muted')

plt.xlabel('Day of the Week')
plt.ylabel('Total Bill')
plt.title('Distribution of Total Bill Amounts by Day and Gender')


plt.show()
```

Create facet gride plot


```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('tips')

g = sns.FacetGrid(data, col="sex", row="time", margin_titles=True)

g.map(sns.scatterplot, "total_bill", "tip", color="blue")


g.fig.suptitle("FacetGrid: Tips vs Total Bill by Gender and Time", y=1.03)


plt.show()
```

creating Heapmap

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data = sns.load_dataset('iris')
correlation_matrix = data.corr(numeric_only=True)

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)

plt.title('Correlation Heatmap of Iris Dataset')
plt.xticks(rotation=45)
plt.yticks(rotation=45)

plt.show()
```
